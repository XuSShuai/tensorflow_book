```python
import tensorflow as tf
import numpy as np
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
DATA_SIZE = 500
dimensions = [2, 10, 50, 10, 1]
n_layers = len(dimensions)
BATCH_SIZE = 32
EPOCH = 5000

data = make_moons(DATA_SIZE, shuffle=True, noise=0.3)
X, Y = data[0], data[1]
X, X_test, Y, Y_test = X[:400], X[400:], np.array(Y[:400]).reshape(-1, 1), np.array(Y[400:]).reshape(-1, 1)
plt.scatter(X[:, 0],X[:, 1], c=np.squeeze(Y.reshape(1, 400)), cmap=plt.cm.Spectral)
plt.show()
```

```python
def get_weight(shape, regularizer, lambd):
    w = tf.Variable(tf.truncated_normal(shape=shape, stddev=0.1))
    if regularizer:
        tf.add_to_collection("losses", tf.contrib.layers.l2_regularizer(lambd)(w))
    return w

def get_bias(shape):
    return tf.Variable(tf.zeros(shape=shape))
```

```python
x = tf.placeholder(shape=[None, 2], dtype=tf.float32)
y = tf.placeholder(shape=[None, 1], dtype=tf.float32)

current_layer = x
for d in range(n_layers - 1):
    weight = get_weight([dimensions[d], dimensions[d + 1]], True, 0.001)
    bias = get_bias([dimensions[d + 1]])
    if d == n_layers - 2:
        current_layer = tf.matmul(current_layer, weight) + bias
    else:
        current_layer = tf.nn.relu(tf.matmul(current_layer, weight) + bias)

global_step = tf.Variable(0)
learning_rate = tf.train.exponential_decay(0.01, global_step, EPOCH/BATCH_SIZE, 0.9, True)

loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=y, logits=current_layer)) + tf.add_n(tf.get_collection("losses"))
train = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)
accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.cast(tf.sigmoid(current_layer) > 0.5, tf.float32), y), tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(EPOCH + 1):
        start = (i * BATCH_SIZE) % DATA_SIZE
        end = min(start + BATCH_SIZE, DATA_SIZE)
        sess.run(train, feed_dict={x: X[start: end], y: Y[start: end]})
        if i % 500 == 0:
            loss_train, acc_train = sess.run([loss, accuracy], feed_dict={x: X, y: Y})
            loss_test, acc_test = sess.run([loss, accuracy], feed_dict={x: X_test, y: Y_test})
            print("After %5d epoch, the train loss is %f, test loss is %f, "
                  "accruacy of train is %f, accuracy of test is %f" % (i, loss_train, loss_test, acc_train, acc_test))
```
